```python
import colorblind_library_copy
import cv2
import matplotlib.pyplot as plt

#cargar aqui imagen que se quiere simular
load_image = cv2.imread('4.jpg')
load_image = cv2.cvtColor(load_image,cv2.COLOR_BGR2RGB)
plt.imshow(load_image)
```

### Protanopia

```python
simulated_img_p = colorblind_library_copy.simulate_colorblindness(load_image, colorblind_type = 'protanopia')
plt.imshow(simulated_img_p)
```

### Deuteranopia

```python
simulated_img_d = colorblind_library_copy.simulate_colorblindness(load_image, colorblind_type = 'deuteranopia')
plt.imshow(simulated_img_d)
```

### Tritanopia

```python
simulated_img_t = colorblind_library_copy.simulate_colorblindness(load_image, colorblind_type = 'Tritanopia')
plt.imshow(simulated_img_t)
```

## Corrección de imágenes por HSV



```python
hsv_corrected = colorblind_library_copy.hsv_color_correct(simulated_img_p, colorblind_type = 'protanopia')

plt.imshow(hsv_corrected)
```

```python
hsv_corrected = colorblind_library_copy.hsv_color_correct(simulated_img_d, colorblind_type='deuteranopia')

plt.imshow(hsv_corrected)
```

```python
hsv_corrected = colorblind_library_copy.hsv_color_correct(simulated_img_t, colorblind_type='tritanopia')

plt.imshow(hsv_corrected)
```

```python
daltonized_img = colorblind_library_copy.daltonize_correct(simulated_img_p, colorblind_type='protanopia')
plt.imshow(daltonized_img)
```

```python
daltonized_img = colorblind_library_copy.daltonize_correct(simulated_img_d, colorblind_type='deuteranopia')
plt.imshow(daltonized_img)
```
